```python
from transformers import AutoProcessor, LlavaOnevisionForConditionalGeneration
import torch

processor = AutoProcessor.from_pretrained("llava-hf/llava-onevision-qwen2-7b-ov-hf")
model = LlavaOnevisionForConditionalGeneration.from_pretrained(
    "llava-hf/llava-onevision-qwen2-7b-ov-hf",
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True,
    device_map="cuda:0"
)
```

## count 채우기

```python
import torch
import math
import pandas as pd
import os
from PIL import Image

def run_vlm_counting(df, index):
    base_path = "YOUR_BASE_PATH"
    comparison_column = "file_path"
    output_path = f'YOUR_OUTPUT_PATH/with_count_{index}.csv'

    global model, processor

    count = []

    for i in range(len(df)):
        object_name = df['object'][i]
        image_path = str(df[comparison_column][i])
        image = Image.open(os.path.join(base_path, image_path)).convert("RGB")

        conversation = [
            {
                "role": "user",
                "content": [
                    {"type": "image", "image": image},
                    {"type": "text", "text": f"How many {object_name} are in the image? Reply with only a number"},
                ],
            },
        ]

        inputs = processor.apply_chat_template(
            conversation,
            add_generation_prompt=True,
            tokenize=True,
            return_dict=True,
            return_tensors="pt"
        )
        inputs = inputs.to("cuda:0", torch.float16)

        output = model.generate(**inputs, max_new_tokens=20)
        full_response = processor.decode(output[0], skip_special_tokens=True)

        if "assistant\n" in full_response:
            answer = full_response.split("assistant\n")[-1].strip()
        else:
            answer = full_response.strip()

        count.append(answer)

    def convert_to_number(value):
        eng_to_number = {
            "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
            "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10,
        }
        if value is None or (isinstance(value, float) and math.isnan(value)):
            return 0
        value = str(value).strip().lower()
        if value.isdigit():
            return int(value)
        if value in eng_to_number:
            return eng_to_number[value]
        return 0

    df = df.copy()
    df['count_vlm'] = [convert_to_number(x) for x in count]
    df.to_csv(output_path, index=False)
    return df
```

```python
import pandas as pd
file_path = 'YOUR_IMAGE_BASELINE_FILE_PATH'
file_path2 = 'YOUR_IMAGE_MULTIPLE_FILE_PATH'
file_path3 = 'YOUR_IMAGE_SINGLE_FILE_PATH'

df = pd.read_excel(file_path)
df2 = pd.read_excel(file_path2)
df3 = pd.read_excel(file_path3)

run_vlm_counting(df,1)
run_vlm_counting(df2,2)
run_vlm_counting(df3,3)
```

## Metric
- acc
- mae
- rmse

```python
import numpy as np
def compute_acc(target : list, count : list)->float:
    """
    Accuracy
    """
    total = 0
    for i in range(len(count)):
      if target[i] == count[i] :
        total+=1

    return total/len(count)

def compute_mae(target, count):
    """
    Mean Absolute Error
    """
    target = np.array(target)
    count = np.array(count)
    return np.mean(np.abs(count - target))

def compute_rmse(target, count):
    """
    Root Mean Squared Error
    """
    target = np.array(target)
    count = np.array(count)
    return np.sqrt(np.mean((count - target) ** 2))
```

```python

df = pd.read_csv("YOUR_COUNT_BASELINE_FILE_PATH")
df2 = pd.read_csv("YOUR_COUNT_MULTIPLE_FILE_PATH")
df3= pd.read_csv("YOUR_COUNT_SINGLE_FILE_PATH")
target1 = df['count']
count1 = df['count_vlm']
prompt1 = df['prompt']


print("our models ver baseline")
print(compute_acc(target1, count1))
print(compute_mae(target1, count1))
print(compute_rmse(target1, count1))
print()


target2 = df2['count']
count2 = df2['count_vlm']
prompt2 = df2['prompt']

print("our models ver multiple")
print(compute_acc(target2, count2))
print(compute_mae(target2, count2))
print(compute_rmse(target2, count2))
print()

target3 = df3['count']
count3 = df3['count_vlm']
prompt3 = df3['prompt']

print("our models ver single")
print(compute_acc(target3, count3))
print(compute_mae(target3, count3))
print(compute_rmse(target3, count3))
print()
```
